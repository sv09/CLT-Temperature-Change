```python
import pandas as pd
import numpy as np
import math
```

```python
data = pd.read_csv("monthly_avg_clt_temp_1878_Present.csv")
```

```python
data.head()
```

```python
for i in range(len(data)):
    year = data.loc[i, "YEAR"]
    if year%10 == 0:
        decade = year
    data.loc[i, "Decade"] = decade
```

```python
data = data.drop(columns={'ANNUAL'})
```

```python
data
```

```python
data = data.groupby('Decade', as_index=False)['JAN', 'FEB', 'MAR', 'APR', 'MAY', 'JUN', 'JUL', 'AUG', 'SEP', 'OCT', 'NOV', 'DEC'].mean()
```

```python
data
```

```python
data_pct_change = data.loc[:, data.columns != 'Decade'].pct_change()
```

```python
data_pct_change 
```

```python
data_pct_change = data_pct_change.rename(columns={'JAN': 'Jan_pct_change', 'FEB':'Feb_pct_change', 'MAR':'Mar_pct_change', 'APR':'Apr_pct_change', 'MAY':'May_pct_change', 'JUN':'Jun_pct_change', 'JUL':'Jul_pct_change', 'AUG':'Aug_pct_change', 'SEP':'Sep_pct_change', 'OCT':'Oct_pct_change', 'NOV':'Nov_pct_change', 'DEC':'Dec_pct_change'})
```

```python
data_pct_change
```

```python
decade_data = pd.concat([data, data_pct_change], axis=1)
```

```python
data.columns
```

```python
decade_data
```

```python
decade_data = decade_data.rename(columns={'JAN':'Jan', 'FEB':'Feb', 'MAR':'Mar', 'APR':'Apr', 'MAY':'May', 'JUN':'Jun', 'JUL':'Jul', 'AUG':'Aug', 'SEP':'Sep', 'OCT':'Oct', 'NOV':'Nov', 'DEC':'Dec'})
```

```python
decade_data
```

```python
decade_data.to_csv("clt_avg_temp_decade.csv", index=False)
```
